```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(1,3,sharex=True,sharey=True,squeeze=False,figsize=(21,8))
ax[0][0].plot(x,tan,color='blue',label='tanx')
ax[0][1].plot(x,sin,color='red',label='sinx')
ax[0][2].plot(x,cos,color='green',label='cosx')

fig.legend(loc='upper center',fontsize=10)
ax[0][0].set_title('Grafik tanx')
ax[0][1].set_title('Grafik sinx')
ax[0][2].set_title('Grafik cosx')
plt.show()
```
